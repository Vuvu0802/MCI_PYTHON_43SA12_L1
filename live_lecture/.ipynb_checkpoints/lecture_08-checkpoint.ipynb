```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

- Figure: 1 khung nh
- Axes: tm nh

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin func', linestyle='--')
plt.plot(x, np.cos(x), label='cos func', linestyle='-.')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Sin function')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

```python
tip_df = sns.load_dataset('tips')
tip_df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(tip_df.total_bill, tip_df.tip, color='g', marker='*')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.grid(True)
plt.title('Total bill vs tip');
```

```python
tip_df.groupby('sex')[['total_bill', 'tip']].sum().T.plot(kind='bar', figsize=(10, 6));
```

```python
gender_cost = tip_df.groupby('sex')[['total_bill', 'tip']].sum().T
gender_cost
```

```python
plt.bar(gender_cost.index[0], gender_cost['Male'], label='Male')
plt.bar(gender_cost.index[1], gender_cost['Female'], label='Female');
```

```python
tip_df.groupby('sex')[['total_bill', 'tip']].sum().T.plot(kind='barh', figsize=(10, 6));
```

- Structural data: 
- Unstructural data:

- Continuous: measurable, [0, 1] --> 0.0001, 0.00021324
- Discrete: countable, [1, 40] --> limit, days of week, weeks of month

- Ordinal: date in month, year of birth, position, 
- Normial: gender, season,...

```python
x = np.random.normal(2, 2, 10000)
plt.figure(figsize=(10, 6))
plt.hist(x, bins=100)
plt.xlabel('Value of column')
plt.ylabel('Frequency of values');
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(tip_df.total_bill);
```

```python
tip_df.groupby('day').total_bill.sum().plot(kind='pie', autopct='%.2f%%', explode=[0, 0.4, 0, 0]);
```

```python
pct_days = tip_df.groupby('day').total_bill.sum()
plt.pie(pct_days.values, [0, 0.4, 0, 0], pct_days.index, autopct='%.2f%%', shadow=True);
```

```python
gender_cost = tip_df.groupby('sex')[['total_bill', 'tip']].sum()
gender_cost
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.pie(pct_days.values, [0, 0.4, 0, 0], pct_days.index, autopct='%.2f%%', shadow=True)
plt.title('Pie chart')

plt.subplot(1, 2, 2)
plt.bar(gender_cost.index, gender_cost.total_bill)
plt.title('bar chart')

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 6))
ax[1].bar(gender_cost.index, gender_cost.total_bill)
ax[0].pie(pct_days.values, [0, 0.4, 0, 0], pct_days.index, autopct='%.2f%%', shadow=True)
tip_df.groupby('sex')[['total_bill', 'tip']].sum().T.plot(kind='bar', ax=ax[2]);
plt.show()
```

```python
sns.scatterplot(data=tip_df, x='total_bill', y='tip', hue='smoker')
plt.title('Total bill vs tip');
```

```python
rs = np.random.RandomState(365)
values = rs.randn(365, 4).cumsum(axis=0)
dates = pd.date_range("1 1 2016", periods=365, freq="D")
data = pd.DataFrame(values, dates, columns=["A", "B", "C", "D"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
plt.show()
```
